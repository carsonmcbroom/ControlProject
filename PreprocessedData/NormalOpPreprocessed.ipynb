```python
import pandas as pd
#load data from csv and print to confirm import
df = pd.read_csv (r'C:\Users\nflfa\Desktop\Git\ControlProject\RawData\Know_Normal_Operation.csv')
print (df)
```

```python
#fill empty values with attribute mean
df = df.fillna(df.mean())
print(df)
```

```python
import sklearn
from sklearn.model_selection import train_test_split as tts

y = df["XMEAS1"]
x = df.drop(columns = "XMEAS1")

print(y)
print(x)
```

```python
x_train, x_test, y_train, y_test = tts(x, y, test_size=0.2, random_state=42)
```

```python
traindf = pd.DataFrame(x_train, y_train)
traindf = df.drop(columns ="Index")
print(traindf)
```

```python
#scale data to be centered around 0
centered_df = traindf.apply(lambda V1: V1-V1.mean())
print(centered_df)
```
